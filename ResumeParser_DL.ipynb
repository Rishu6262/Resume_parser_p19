```python
import fitz  # PyMuPDF

doc = fitz.open("rishu gurjar.docx")
text = ""

for page in doc:
    text += page.get_text()

print(text)
```

```python
import re
email = re.findall(r'\S+@\S+', text)
phone = re.findall(r'\d{10}', text)

print(email,phone)
```

```python
skills = ["python", "java", "sql", "machine learning",
    "deep learning", "pandas", "numpy", "tensorflow",
    "pytorch", "excel"]

found_skills = []
for skill in skills:
    if skill in text.lower() :
        found_skills.append(skill)

print(found_skills)
```
